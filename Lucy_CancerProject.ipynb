# Loading in data from Scikit-Learn

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

df =  pd.DataFrame(cancer.data, columns = cancer.feature_names)
df["target"] = cancer.target
df
```

# Running a logistic regression to identify if certain variables can be used to predict cancer accurately

Model development and prediction

```python
from sklearn.linear_model import LogisticRegression

X = df[["mean radius", "mean perimeter", "mean area"]]
y = df["target"]

model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split

# Using 30% of the dataset to train the model
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3)
```

```python
# Fit the model 
model.fit(train_X, train_y)
```

```python
cancer_pred = model.predict(test_X)
```

Evaluating the model using Confusion Matrix 

```python
from sklearn import metrics

cnf_matrix = metrics.confusion_matrix(test_y, cancer_pred)
cnf_matrix

# The confusion matrix is returned in the form of the array object
# The dimension of the matrix is 2*2 because this model is binary classification i.e. 0 or 1 
```

Diagonal values represent accurate predictions and non-diagonal elements are inaccurate predictions

From this model prediction we can see there are fairly few incorrect predictions but this model requires validation on larger datasets

```python
model.score(test_X, test_y)
```

We can see that the logistic regression model scores quite reasonably but could be improved

# Visualising the results of the model as a confusion matrix using matplotlib and seaborn

```python
# The confusion matrix is visualised using Heatmap

class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)

# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

Evaluating the model using classification_report for accuracy, precision, and recall

```python
from sklearn.metrics import classification_report

print(classification_report(test_y, cancer_pred))
```

```python
# Defining the classifications to make it accessible and understandable
target_names = ['with cancer', 'without cancer']

print(classification_report(test_y, cancer_pred, target_names=target_names))
```

The classification rate suggests good model accuracy

The model is overall slightly more precise for predicting benign tumours compared to cancerous tumours
